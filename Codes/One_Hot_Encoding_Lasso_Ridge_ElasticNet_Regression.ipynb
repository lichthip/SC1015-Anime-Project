```python
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt 
sb.set()
```

```python
anime_clean=pd.read_csv('anime_clean.csv')
anime_clean
```

```python
# Perform one-hot encoding
genres_encoded = anime_clean['genre'].str.get_dummies(sep=', ')

# Join one-hot encoded genres with the anime_clean_encoded DataFrame
anime_clean_encoded = pd.concat([anime_clean, genres_encoded], axis=1)

# Drop the original 'genre' column (optional)
anime_clean_encoded.drop(columns=['genre'], inplace=True)
anime_clean_encoded
```

```python
anime_clean_encoded = pd.get_dummies(anime_clean_encoded, columns=['type'], prefix='', prefix_sep='')
anime_clean_encoded
```

```python
# Convert True/False to 1/0 only for specific columns
columns_to_convert = ['Movie', 'TV', 'OVA', 'Others']
for column in columns_to_convert:
    anime_clean_encoded[column] = anime_clean_encoded[column].astype(int)

anime_clean_encoded
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Lasso, Ridge, ElasticNet
from sklearn.model_selection import GridSearchCV, cross_val_score
x = anime_clean_encoded.drop(columns=['name', 'rating', 'anime_id'])
y = anime_clean_encoded['rating']  # Target variable
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=26)

#Define the parameter grid for Lasso, Ridge, and ElasticNet
lasso_param_grid = {'alpha': [0.001, 0.01, 0.1, 1, 10]}
ridge_param_grid = {'alpha': [0.001, 0.01, 0.1, 1, 10]}
elastic_net_param_grid = {'alpha': [0.001, 0.01, 0.1, 1, 10],
                         'l1_ratio': [0.1, 0.3, 0.5, 0.7, 0.9]}

# Instantiate Lasso, Ridge, and ElasticNet models
lasso_model = Lasso()
ridge_model = Ridge()
elastic_net_model = ElasticNet()

# Perform GridSearchCV for Lasso
lasso_grid_search = GridSearchCV(lasso_model, param_grid=lasso_param_grid, cv=5, scoring='neg_mean_squared_error')
lasso_grid_search.fit(x_train, y_train)

# Perform GridSearchCV for Ridge
ridge_grid_search = GridSearchCV(ridge_model, param_grid=ridge_param_grid, cv=5, scoring='neg_mean_squared_error')
ridge_grid_search.fit(x_train, y_train)

# Perform GridSearchCV for ElasticNet
elastic_net_grid_search = GridSearchCV(elastic_net_model, param_grid=elastic_net_param_grid, cv=5, scoring='neg_mean_squared_error')
elastic_net_grid_search.fit(x_train, y_train)

# Get best parameters and best scores for Lasso
best_params_lasso = lasso_grid_search.best_params_
best_score_lasso = lasso_grid_search.best_score_

# Get best parameters and best scores for Ridge
best_params_ridge = ridge_grid_search.best_params_
best_score_ridge = ridge_grid_search.best_score_

# Get best parameters and best scores for ElasticNet
best_params_elastic_net = elastic_net_grid_search.best_params_
best_score_elastic_net = elastic_net_grid_search.best_score_

# Fit models with best parameters
lasso_model_best = Lasso(alpha=best_params_lasso['alpha'])
ridge_model_best = Ridge(alpha=best_params_ridge['alpha'])
elastic_net_model_best = ElasticNet(alpha=best_params_elastic_net['alpha'], l1_ratio=best_params_elastic_net['l1_ratio'])

lasso_model_best.fit(x_train, y_train)
ridge_model_best.fit(x_train, y_train)
elastic_net_model_best.fit(x_train, y_train)

# Predictions
y_pred_lasso = lasso_model_best.predict(x_test)
y_pred_ridge = ridge_model_best.predict(x_test)
y_pred_elastic_net = elastic_net_model_best.predict(x_test)

# Calculate R² and adjusted R² for Lasso
lasso_r_squared = lasso_model_best.score(x_test, y_test)
n_test = len(y_test)
p_test = x_test.shape[1]
lasso_adjusted_r_squared = 1 - (1 - lasso_r_squared) * (n_test - 1) / (n_test - p_test - 1)

# Calculate R² and adjusted R² for Ridge
ridge_r_squared = ridge_model_best.score(x_test, y_test)
ridge_adjusted_r_squared = 1 - (1 - ridge_r_squared) * (n_test - 1) / (n_test - p_test - 1)

# Calculate R² and adjusted R² for ElasticNet
elastic_net_r_squared = elastic_net_model_best.score(x_test, y_test)
elastic_net_adjusted_r_squared = 1 - (1 - elastic_net_r_squared) * (n_test - 1) / (n_test - p_test - 1)

# Print results
print("Lasso Best Parameters:", best_params_lasso)
print("Lasso Best Score (Negative MSE):", -best_score_lasso)
print("Lasso R²:", lasso_r_squared)
print("Lasso Adjusted R²:", lasso_adjusted_r_squared)

print("Ridge Best Parameters:", best_params_ridge)
print("Ridge Best Score (Negative MSE):", -best_score_ridge)
print("Ridge R²:", ridge_r_squared)
print("Ridge Adjusted R²:", ridge_adjusted_r_squared)

print("ElasticNet Best Parameters:", best_params_elastic_net)
print("ElasticNet Best Score (Negative MSE):", -best_score_elastic_net)
print("ElasticNet R²:", elastic_net_r_squared)
print("ElasticNet Adjusted R²:", elastic_net_adjusted_r_squared)
```

```python
# Create a DataFrame to store coefficients and variable names for Lasso model
lasso_coefficients_df = pd.DataFrame({'Variable': x_train.columns, 'Coefficient': lasso_model_best.coef_})
lasso_coefficients_df = lasso_coefficients_df.sort_values(by='Coefficient', ascending=False)

# Print coefficients and variable names for Lasso model
print("Lasso Coefficients:")
print(lasso_coefficients_df)

# Create a DataFrame to store coefficients and variable names for Ridge model
ridge_coefficients_df = pd.DataFrame({'Variable': x_train.columns, 'Coefficient': ridge_model_best.coef_})
ridge_coefficients_df = ridge_coefficients_df.sort_values(by='Coefficient', ascending=False)

# Print coefficients and variable names for Ridge model
print("\nRidge Coefficients:")
print(ridge_coefficients_df)

# Create a DataFrame to store coefficients and variable names for ElasticNet model
elastic_net_coefficients_df = pd.DataFrame({'Variable': x_train.columns, 'Coefficient': elastic_net_model_best.coef_})
elastic_net_coefficients_df = elastic_net_coefficients_df.sort_values(by='Coefficient', ascending=False)

# Print coefficients and variable names for ElasticNet model
print("\nElasticNet Coefficients:")
print(elastic_net_coefficients_df)
```
